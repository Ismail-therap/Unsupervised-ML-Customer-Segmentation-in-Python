```python
# Hiarchical Clustering
data = df.iloc[:, 3:5].values 

plt.figure(figsize=(10,10))  
plt.title("Customer Dendograms")  
dend = shc.dendrogram(shc.linkage(data, method='ward')) 
```

```python
cluster = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='ward')  
cluster.fit_predict(data)  
```

```python
plt.figure(figsize=(10, 7))  
plt.scatter(data[:,0], data[:,1], c=cluster.labels_, cmap='rainbow') 
```

```python
# K-means clustering

# Loading the data
df = pd.read_csv('C:\\Users\\Saiful_desv\\Desktop\\Segmentation in Python\\segmentation_draft.csv')

df[['totalcomb', 'newrfpdtq','combnewrenrfpsold']] = StandardScaler().fit_transform(df[['totalcomb', 'newrfpdtq','combnewrenrfpsold']])
#data = df.iloc[:,3:5].values 



clmns = ['totalcomb', 'newrfpdtq','combnewrenrfpsold']
data = df[clmns]



# from sklearn.cluster import KMeans 
clusters = 4
  
kmeans = KMeans(n_clusters = clusters) 
kmeans.fit(data) 

labels = kmeans.labels_

#Glue back to originaal data
data['clusters'] = labels
#Add the column into our list
clmns.extend(['clusters'])
print (data[clmns].groupby(['clusters']).mean())
```

```python
#Scatter plot of Wattage and Duration
plt.figure(figsize=(10, 7))  

sns.lmplot('totalcomb', 'newrfpdtq', 
           data=data, 
           fit_reg=False, 
           hue="clusters",  
           scatter_kws={"marker": "D", 
                        "s": 100})

plt.title('totalcomb vs newrfpdtq')
plt.xlabel('newrfpdtq')
plt.ylabel('newrfpdtq')
```

```python
# Gaussian Mixed Model

# Loading the data
df = pd.read_csv('C:\\Users\\Saiful_desv\\Desktop\\Segmentation in Python\\segmentation_draft.csv')
df[['totalcomb', 'newrfpdtq','combnewrenrfpsold']] = StandardScaler().fit_transform(df[['totalcomb', 'newrfpdtq','combnewrenrfpsold']])

X = df.iloc[:,3:5].values 
```

```python
gmm = GMM(n_components=4).fit(X)
labels = gmm.predict(X)
```

```python
# plot the cluster assignments
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.xlabel("totalcomb")
plt.ylabel("newrfpdtq")
```

```python
# DBSCANE 

# cluster the data into five clusters
dbscan = DBSCAN(eps=0.123, min_samples = 2)
clusters = dbscan.fit_predict(X)
```

```python
# plot the cluster assignments
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap='viridis')
plt.xlabel("totalcomb")
plt.ylabel("newrfpdtq")
```
